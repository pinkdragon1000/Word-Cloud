```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
from wordcloud import WordCloud, STOPWORDS
import random
```

```python
def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(70, 100)
```

```python
p = os.getcwd()

# Read the whole text.
text = open(path.join(p, 'sherlockholmes.txt')).read()

# read the mask image
mask = np.array(Image.open(path.join(p, "sherlock.jpg")))

stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color="black", mask=mask, stopwords=stopwords, random_state=1)
# generate word cloud
wc.generate(text)


%matplotlib inline
plt.suptitle("Sherlock Holmes Word Cloud", fontsize=30)
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=3), interpolation="bilinear")

plt.axis("off")
plt.figure()
plt.axis("off")
plt.show()
```
