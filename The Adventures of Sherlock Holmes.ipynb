Word Cloud-Sherlock Holmes

```python
import os
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
os.chdir(r"/home/sita/Desktop")
text = open('sherlockholmes.txt', "r")

data=text.read().lower()
data

%matplotlib inline
plt.figure(figsize=(20,10))

wordcloud = WordCloud(background_color='white', mode = "RGB", width = 2000, height=1000, collocations=False).generate(data)
plt.suptitle("Sherlock Holmes Word Cloud", fontsize=50)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
